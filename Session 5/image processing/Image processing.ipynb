## Image processing.
library image processing.

```python
import cv2
img=cv2.imread('owl.jpeg')
img
```

```python
img.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

## cv2 consider image as BGR

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2
```

```python
plt.imshow(img2)
```

## Split above in rgb components

```python
r ,g ,b = cv2.split(img2)
```

```python
r
```

```python
r.shape
```

```python
g
```

```python
g.shape
```

```python
b
```

```python
b.shape
```

```python
plt.imshow(r,cmap='Reds')
```

```python
plt.imshow(g,cmap='Greens')
```

```python
plt.imshow(b,cmap='Blues')
```

## Reading image as gray scale

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_gray
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

## Reading Gray scaling directly

```python
import seaborn as sns
plt.figure(figsize=(12,12))
sns.heatmap(img_gray,annot=True,fmt='d')
plt.show()
```
